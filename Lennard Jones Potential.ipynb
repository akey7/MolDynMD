# Lennard Jones Potential

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lennard_jones_potential(r_angstroms, epsilon, r0_angstroms):
    """
    This calculates the Lennard Jones potential of a bond length.
    
    Parameters
    ----------
    r_angstroms: float or np.array
        Length(s) of the bond(s) in Angstroms.
        
    epsilon : float
        Depth of potential well. In kcal A^6 / mol.
    
    r0_angstroms: float
        The equilibrium bond length in the potential. Angstroms.
        Defaults to a midpoint for the H2 molecule.
        
    Returns
    -------
    np.array
        Array of floats of potential energies. kcal/mol
        
    """
    repulsive = (r0_angstroms / r_angstroms) ** 12
    attractive = -2 * (r0_angstroms / r_angstroms) ** 6
    vlj = epsilon * (repulsive + attractive)
    return vlj
```

```python
r_start_angstroms = 2.3
r_end_angstroms = 3
r0_angstroms = 2.5
epsilon = 0.2

radii_angstroms = np.linspace(r_start_angstroms, r_end_angstroms, 100)
potentials_kcal_mol = lennard_jones_potential(radii_angstroms, epsilon, r0_angstroms)

x_min = radii_angstroms.min()
x_max = radii_angstroms.max()
y_min = potentials_kcal_mol.min() * 1.1
y_max = potentials_kcal_mol.max()

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.set_title(f"Lennard-Jones Potential, r0={r0_angstroms} A, epsilon={epsilon}")
ax.set_xlabel("r")
ax.set_ylabel("V")
ax.set_ylim(y_min, y_max)
ax.set_xlim(x_min, x_max)
ax.plot(radii_angstroms, potentials_kcal_mol)
ax.axhline(lw=1, color='r')
ax.axvline(r0_angstroms, lw=1, color='r')
```
