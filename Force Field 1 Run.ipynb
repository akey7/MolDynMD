# MolDyn, 2 Atoms

```python
import numpy as np
import matplotlib.pyplot as plt
from force_field_1 import MolDynTwoAtom
%matplotlib inline
```

## Run 1

\\(\epsilon = 0.2 \\), so the potential energy well isn't too deep.

```python
mol_dyn = MolDynTwoAtom(r0_A=2.5,
                        epsilon=0.2,
                         r1_A=0,
                         r2_A=5,
                         v1_A_per_fs=0.01,
                         v2_A_per_fs=-0.01,
                         dt_fs=0.1,
                         grad_delta_A=0.01)
for _ in range(1000):
    mol_dyn.timestep()
    
_, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

xs = np.array(mol_dyn.timesteps)
axs[0].plot(xs, np.array(mol_dyn.r1s), label="atom 1")
axs[0].plot(xs, np.array(mol_dyn.r2s), label="atom 2")
axs[0].set_xlim(xs.min(), xs.max())
axs[0].legend(loc="best")
axs[0].set_title("Positions")
axs[0].set_xlabel("Time (fs)")
axs[0].set_ylabel("x (Angstroms)")

axs[1].plot(xs, np.array(mol_dyn.v1s), label="atom 1")
axs[1].plot(xs, np.array(mol_dyn.v2s), label="atom 2")
axs[1].set_xlim(xs.min(), xs.max())
axs[1].legend(loc="best")
axs[1].set_title("Velocities")
axs[1].set_xlabel("Time (fs)")
axs[1].set_ylabel("v (Angstroms / fs)")
```
